```python
url = "https://public.fyers.in/sym_details/NSE_FO.csv"

fno_symbolList = pd.read_csv(url, header=None, usecols=range(21))

# Explicitly named columns
named_columns = ['FyersToken', 'Name', 'Instruemnt', 'lot', 'tick', 'ISIN', 'TradingSession', 'Lastupdated', 'Expirydate', 'Symbol'] 

# Assign names to the first 10 columns
fno_symbolList.columns = named_columns + list(range(21-len(named_columns)))  

fno_symbolList['Expirydate'] = pd.to_datetime(fno_symbolList['Expirydate'], unit='s').apply(lambda x:x.date())

fno_symbolList.to_csv("FO.csv", index=False)
```

```python
# Load the CSV
fno_symbolList = pd.read_csv("FO.csv")

# Filter rows where 'Symbol' contains 'NSE:NIFTY24912'
filtered_df = fno_symbolList[fno_symbolList['Symbol'].str.contains("NSE:NIFTY24SEP", na=False)]

# Iterate through the filtered DataFrame and print each row in key-value format
for index, row in filtered_df.iterrows():
    print(f"Symbol: {row['Symbol']}, Name: {row['Name']}")


# Select the 'Symbol' and 'Name' columns for saving
output_df = filtered_df[['Symbol', 'Name']]

# Convert the filtered DataFrame to CSV
output_df.to_csv('filtered_FO_symbols.csv', index=False)
```
